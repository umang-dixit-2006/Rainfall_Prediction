# Rainfall Detection Project

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('../austin_weather.csv')
data
```

```python
# Data cleaning
data= data.drop(["Events", "Date", "SeaLevelPressureLowInches"], axis=1)
```

```python
data = data.replace("T", "0.0")
```

```python
data = data.replace("-", "0.0")
```

```python
data.to_csv('../austin_weather_cleaned.csv', index=False)
```

```python
data = pd.read_csv('../austin_weather_cleaned.csv')
data
```

```python
X = data.drop(['PrecipitationSumInches'], axis=1)
```

```python
Y = data['PrecipitationSumInches']
```

```python
# reshaping it into 2d vector
Y = Y.values.reshape(-1,1)
```

```python
day_index = 768
days = [i for i in range(Y.size)]
days
```

```python
# intialize the linear regression classifier
clf = LinearRegression()

# train the classifier
clf.fit(X, Y)
```

```python
# plot a graph
print("The Precipitation trend Graph:")
plt.scatter(days, Y, color='blue')
plt.scatter(days[day_index], Y[day_index], color='red', s=100)
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The Precipitation vs Attributes Graph:")

for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]], color = 'green')
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color = 'red')
    plt.title(x_vis.columns.values[i])
plt.show()
```
